# Exploring Future Value of Money

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
import ipywidgets as widgets
from IPython.display import display
%matplotlib inline

def show_fv(rate=0.08, years=25, starting=1000):
    plt.figure(figsize=(10,8))
    y = [npf.fv(rate=rate, nper=i, pmt=0, pv=-starting) for i in range(years)]
    
    plt.plot(y)
    plt.ylabel(f'Future value of ${starting}, dollars')
    plt.xlabel('years')
    
controls = widgets.interactive(show_fv, rate=(0.01,0.15, 0.01), years=(1, 65), starting=(1000, 100000, 1000))
display(controls)
```
